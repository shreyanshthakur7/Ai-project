```python
import streamlit as st
import random

st.set_page_config(page_title="MoodVerse AI", page_icon="ðŸŽµ", layout="centered")

# -----------------------------
# Data Store for Poems & Songs
# -----------------------------
content = {
    "poem": {
        "happy": [
            "Sunshine spills across my face,\nEvery shadow fades away in grace.",
            "Laughter blooms like morning dew,\nColors brighter, skies more blue.",
            "Joy hums softly through the air,\nLife feels light without a care."
        ],
        "sad": [
            "Rain taps soft on my windowpane,\nEach drop whispers my hidden pain.",
            "The night feels long, my thoughts unwind,\nLost echoes linger in my mind.",
            "Tears fall quietly in rhyme,\nWounds heal only with time."
        ],
        "angry": [
            "Fire burns behind my eyes,\nTruth cuts deep through all the lies.",
            "Thunder rolls inside my chest,\nFuryâ€™s tide will not be suppressed.",
            "My words are lightning, sharp and bright,\nBreaking silence into fight."
        ],
        "romantic": [
            "Your gaze a star, my heart its sky,\nEach moment with you passes by.",
            "You speak, and time begins to slow,\nIn your warmth, I find my glow.",
            "Love whispers softly, hearts entwine,\nYour soul forever bound with mine."
        ]
    },
    "song": {
        "happy": [
            "ðŸŽ¶ Dancing through sunlight, I feel alive,\nDreams in my pocket, stars in my eyes.",
            "ðŸŽµ Clap your hands, weâ€™re chasing dreams tonight,\nEvery heartbeat shining bright.",
            "ðŸŽ¶ Singing loud, the skyâ€™s so clear,\nHappiness echoes far and near."
        ],
        "sad": [
            "ðŸŽµ Lost in echoes, fading lines,\nTears fall slow like whispered rhymes.",
            "ðŸŽ¶ The rain keeps singing our goodbye,\nBut I still see your shadow in the sky.",
            "ðŸŽµ My heart still hums your melody,\nThough time erased our memory."
        ],
        "angry": [
            "ðŸŽ¶ Iâ€™m breaking chains tonight,\nNo fear left inside the fight.",
            "ðŸŽµ Shout it loud, I wonâ€™t back down,\nMy fire burns, I wear the crown.",
            "ðŸŽ¶ Storms may rage, I stand my ground,\nMy silence screams without a sound."
        ],
        "romantic": [
            "ðŸŽµ Hold me close beneath the moon,\nOur hearts in tune, a timeless tune.",
            "ðŸŽ¶ Every word you sayâ€™s a melody,\nYou and I, weâ€™re harmony.",
            "ðŸŽµ Your smileâ€™s the rhythm, your eyes the song,\nIn your arms I truly belong."
        ]
    }
}

# -----------------------------
# Track Used Creations
# -----------------------------
used = {"poem": {m: [] for m in content["poem"]},
        "song": {m: [] for m in content["song"]}}


# -----------------------------
# Non-Repeating Generator
# -----------------------------
def generate_creation(mood, genre):
    mood = mood.lower()
    genre = genre.lower()
    if mood not in content[genre]:
        return "Mood not recognized. Try happy, sad, angry, or romantic."

    available = [x for x in content[genre][mood] if x not in used[genre][mood]]
    if not available:  # All used, reset
        used[genre][mood].clear()
        available = content[genre][mood]

    choice = random.choice(available)
    used[genre][mood].append(choice)
    return choice


# -----------------------------
# Streamlit UI
# -----------------------------
def main():
    st.title("ðŸŽµ MoodVerse AI")
    st.subheader("AI-Generated Poem | Song Based on Your Mood")

    mood = st.text_input("ðŸ’­ Enter your mood (happy / sad / angry / romantic):")
    genre = st.selectbox("ðŸŽ¶ Choose type:", ["Poem", "Song"])

    if st.button("âœ¨ Generate"):
        if mood.strip() == "":
            st.warning("Please enter a mood ðŸ’¡")
        else:
            with st.spinner("Crafting your creation... ðŸŽ¨"):
                output = generate_creation(mood, genre)
                st.markdown("## âœ¨ Your Generated Creation:")
                st.markdown(f"{output}")

    st.markdown("---")
    st.caption("Â© 2025 United Institute of Technology | Created by Shreyansh Thakur")


# -----------------------------
# Run App
# -----------------------------
if __name__ == "__main__":
    main()
```
